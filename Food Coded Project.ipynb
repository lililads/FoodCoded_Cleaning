# Data Cleaning

## Replace text with NaN in the 'weight' column

```python
df['weight'].replace(["I'm not answering this"], np.nan, inplace=True)
```

## Extract only the numeric values from the 'weight' and 'GPA' columns and create new columns from them

```python
df['weight_clean'] = df['weight'].str.extract(r'(\d+\.?\d*)')
df['GPA_clean'] = df['GPA'].str.extract(r'(\d+\.?\d*)')
```

## Convert 'GPA_clean' and 'weight_clean' columns to float type

```python
df[['GPA_clean', 'weight_clean']] = df[['GPA_clean', 'weight_clean']].astype(float)
```

## Move 'GPA_clean' column to the front of the DataFrame

```python
columna = 'GPA_clean'

resto = [col for col in df.columns if col != columna]

df = df[[columna] + resto]
```

## Drop unnecessary or redundant columns from the DataFrame

```python
df.drop(columns=['comfort_food_reasons_coded', 'eating_changes_coded', 'GPA', 'weight'], inplace=True)
```

## Rename cleaned columns to replace original names

```python
df.rename(columns={
    'GPA_clean': 'GPA',
    'weight_clean': 'weight'
}, inplace=True)
```

## Select numeric columns and convert them to float type

```python
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
df[columnas_numericas] = df[columnas_numericas].astype(float)
```

## Convert all text columns to lowercase

```python
columnas_texto = df.select_dtypes(include=['object']).columns
df[columnas_texto] = df[columnas_texto].apply(lambda col: col.str.lower())
```

## Fill missing values in all columns with the mode of each column

```python
for col in df.columns:
    moda = df[col].mode()
    if not moda.empty:
        df[col].fillna(moda[0], inplace=True)
```
